```python
import urllib

AUDIO_FILENAMES = ['speech_16000_hz_mono.wav', 'speech_48000_hz_mono.wav']

for name in AUDIO_FILENAMES:
  url = f'https://storage.googleapis.com/mediapipe-assets/{name}'
  urllib.request.urlretrieve(url, name)
```
